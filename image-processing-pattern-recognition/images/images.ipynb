# Images

Displays all the input images represented in the `.dat` files. Converts them to `.bmp` and saves them.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_inp = 'dat/'   # path for input files
path_out = 'bmp/'   # path for output files

filenames = [
    'a256',
    'b256',
    'ba256',
    'f256',
    'fu256',
    'l256',
    'newimg',
    'n256',
    'o256',
    'p256',
    'pap256',
    'temp',
    'temp2',
    'upped',
    'z256'
]

ext_inp = '.dat'    # file extention for input
ext_out = '.bmp'    # file extention for output
```

## Display and Save the images as BMP

```python
# Matrix dimensions
cols = 3
rows = -(-len(filenames) // cols)

# Create figure with rows  cols subplots
fig, axs = plt.subplots(rows, cols, dpi=80, sharex=True, sharey=True)
fig.set_size_inches(5 * cols, 5.5 * rows)

# Iterate for all filenames
for idx, filename in enumerate(filenames):
    # Store image pixels as uint8 2D array
    img_matrix = np.array(
        [i.strip().split() for i in open(path_inp + filename + ext_inp).readlines()],
        dtype='uint8'
    )

    # Set subplot title as '"filename" (rows, cols)' 
    axs[int(idx // cols), idx % cols].set_title('"{}" {}'.format(filename + ext_inp, img_matrix.shape))
    # Add subplot to figure plot buffer
    axs[int(idx // cols), idx % cols].imshow(
        img_matrix,
        cmap='gray',
        vmin=0, vmax=255
    )

    # Save image as .bmp file
    plt.imsave(path_out + filename + ext_out, img_matrix, cmap='gray', vmin=0, vmax=255)

# Hide x labels and tick labels for top plots and y ticks for right plots
for ax in axs.flat:
    ax.label_outer()

# Display the figure
plt.show()
```
